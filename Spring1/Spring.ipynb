```python
import pandas as pd
```

```python
datos = pd.read_csv('Spring.csv')
datos
```

```python
def ordenar(columna, orden):
    
    ordenados = datos.sort_values(columna, ascending=False)

    if orden == "ascendente":
        ordenados = datos.sort_values(columna, ascending=True)
        
    return ordenados
```

```python
# Ejemplo de uso
ordenados = ordenar("Edad", "ascendente")
print(ordenados.to_string())
```
